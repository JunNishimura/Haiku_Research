```python
import pandas as pd
import scipy.sparse
from gensim import matutils, models
```

```python
dtm = pd.read_pickle('../../data/pickles/dtm.pickle')
dtm.head()
```

```python
corpus = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(dtm.transpose()))
```

```python
lda = models.LdaModel(corpus=corpus, num_topics=2, passes=10)
lda.print_topics()
```
