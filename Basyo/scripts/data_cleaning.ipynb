### 季語csvの作成

```python
import glob
import sys, os
sys.path.append(os.path.join(os.path.curdir, '../../'))
from common.utils import *
import MeCab
```

```python
def word2season(word):
    # settings
    kigo_titles_jap = dict(zip(kigo_titles, ['春', '夏', '秋', '冬', '新年']))
    with open('../../data/Kigo/kigo_dict.pickle', 'rb') as f:
        kigo_dict = pickle.load(f)
    
    # search
    for kigo in kigo_dict.keys():
        if word in kigo_dict[kigo]:
#             print(word + 'は' + kigo_titles_jap[kigo] + 'の俳句です')
            return kigo_titles_jap[kigo]
    
#     print(word + 'は' + kigo_titles_jap[kigo] + 'の俳句ではありません')
    return None
```

```python
m = MeCab.Tagger('-Ochasen')
haijins = [
    'bashou',
    'buson',
    'issa',
    'siki',
    'souseki',
    'ryunosuke',
    'kyoshi',
    'housai',
    'hekigotoh',
    'hisajo',
]

for haijin in haijins:
    kigo_list = []
    season_list = []
    df = pd.read_csv('../../data/Basyo/%s_raw.csv' % haijin, index_col=0)
    for haiku in df['俳句']:
        morphemes = morphological_analyzer(m, haiku)
        nouns    = extract_words_lexicalCategory('名詞', morphemes)
        for noun in nouns:
            season = word2season(noun)
            if season is not None:
                break
        if season is None:
            season = ''
            noun = ''
        kigo_list.append(noun)
        season_list.append(season)
        
    df['季語'] = kigo_list
    df['季節'] = season_list
    
    df.to_csv('../../data/Basyo/%s.csv' % haijin, encoding='utf_8_sig')
```

## 品詞データフレームの作成

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
haiku_df = pd.read_pickle('../../data/pickles/haiku_df.pickle')
haiku_df.head()
```

```python
names = haiku_df['作者'].unique()
names
```

```python
# pos = part of speech
pos_list = ['名詞', '動詞', '形容詞', '副詞'] 
pos_df = pd.DataFrame(index=names, columns=pos_list)

grouped = haiku_df.groupby('作者')
for name, group in grouped:
    for pos in pos_list:
        pos_df.loc[name, pos] = ' '.join(group[pos].dropna().tolist()) 
```

```python
with open('../../data/pickles/pos_df.pickle', 'wb') as f:
    pickle.dump(pos_df, f)
```
