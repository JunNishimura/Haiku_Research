# 季語が未定の句に季語を割り当てる

```python
import pandas as pd
import pickle
```

```python
# 季語辞書の取得
kigo_dict = pd.read_pickle('../../data/pickles/kigo_dict.pickle')
kigo_dict
```

```python
# 俳句データフレームの取得
df = pd.read_pickle('../../data/pickles/haiku_df.pickle')
df.head()
```

```python
# 季語のない句のみを取り出してdfを作成
no_kigo_df = df[df['季語'].isnull()]
no_kigo_df
```

```python
# 長さ2以上の単語を切り崩して
def word_decomposer(word):
    word_list = []
    for i in range(1, len(word)): # 切り出す文字の長さを格納する変数i
        for j in range(0, len(word)): # ループカウンタ
            if j + i > len(word):
                break
            word_list.append(word[j:j+i+1])
    
    return word_list


def search_season(df_id, noun_list):
    for noun in noun_list.split( ):
        # 長さ2以上の単語は未調査だから分割して全+組み合わせを調べて、季語に合うかを調べる
        if len(noun) >= 2:
            decomposed_words = word_decomposer(noun)
            
            found = False
            for w in decomposed_words:
                for k, v in kigo_dict.items():
                    if w in v:
                        df.loc[df_id, '季節'] = k
                        df.loc[df_id, '季語'] = w
                        found = True
                        break
                # 見つけたらループ抜ける
                if found:
                    break                    
```

```python
for df_id, noun_li in zip(no_kigo_df.index, no_kigo_df['名詞']):
    if type(noun_li) is str:
        search_season(df_id, noun_li)
```
