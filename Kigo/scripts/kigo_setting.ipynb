# 季語辞書の作成

```python
kigo_dict = {}
seasons = ['spring', 'summer', 'fall', 'winter', 'newyear']

for season in seasons:
    kigo_df = pd.read_csv("../../data/Kigo/kigo_{}.csv".format(season), index_col=0)
    title = kigo_df.columns[0]
    kigo_dict[title] = kigo_df[title].to_list()
```

# 季語の再設定

```python
# 季語辞書の取得
kigo_dict = pd.read_pickle('../../data/pickles/kigo_dict.pickle')
kigo_dict
```

```python
# 俳句データフレームの取得
df = pd.read_pickle('../../data/pickles/haiku_df.pickle')
df.head()
```

```python
# まずは名詞からマッチする季語を見付ける
def set_kigo(df_id, nouns):
    for noun in nouns.split():
        for k, v in kigo_dict.items():
            if noun in v:
                selected_df.loc[df_id, '季語'] = noun
                selected_df.loc[df_id, '季節'] = k
                return


selected_df = df.copy()
for df_id, nouns in zip(selected_df.index, selected_df['名詞']):
    if type(nouns) is str:
        set_kigo(df_id, nouns)
```

```python
selected_df
```

```python
# 季語のない句のみを取り出してdfを作成
no_kigo_df = selected_df[selected_df['季語'].isnull()]
no_kigo_df
```

```python
# 長さ2以上の単語を切り崩して
def word_decomposer(word):
    word_list = []
    for i in range(1, len(word)): # 切り出す文字の長さを格納する変数i
        for j in range(0, len(word)): # ループカウンタ
            if j + i > len(word):
                break
            word_list.append(word[j:j+i+1])
    
    return word_list


def set_kigo2(df_id, noun_list):
    for noun in noun_list.split( ):
        # 長さ2以上の単語は未調査だから分割して全+組み合わせを調べて、季語に合うかを調べる
        if len(noun) >= 2:
            decomposed_words = word_decomposer(noun)
            
            for w in decomposed_words:
                for k, v in kigo_dict.items():
                    if w in v:
                        selected_df.loc[df_id, '季節'] = k
                        selected_df.loc[df_id, '季語'] = w
                        return                
```

```python
# 文字列が2以上の単語を切り出して季語探索を図る
for df_id, noun_li in zip(no_kigo_df.index, no_kigo_df['名詞']):
    if type(noun_li) is str:
        set_kigo2(df_id, noun_li)
```
