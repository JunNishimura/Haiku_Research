## Exploratory Data Analysis

### What are some ways you can think of to explore our data set?

- top words
- vocabulary 

```python
import pandas as pd 

data = pd.read_pickle('../pickles/dtm.pkl')
data = data.transpose()
data.head()
```

```python
top_dict = {}
for haijin in data.columns:
    top = data[haijin].sort_values(ascending=False).head(30)
    top_dict[haijin] = list(zip(top.index, top.values))
top_dict
```

```python
for haijin, top_words in top_dict.items():
    print(haijin)
    print(', '.join([word for word, count in top_words[:15]]))
    print('----------')
```

```python
from wordcloud import WordCloud

wc = WordCloud(font_path='HGRGM.TTC', background_color='white', colormap='Dark2', max_font_size=150, random_state=42)
```

```python
corpus = pd.read_pickle('../pickles/corpus.pkl')
corpus.head()
```

```python
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams["font.family"] = "IPAexGothic"

full_names = [
    '松尾芭蕉',
    '宝井其角',
    '服部嵐雪',
    '向井去来',
    '内藤丈草',
    '森川許六',
    '杉山杉風',
    '各務支考',
    '立花北枝',
    '志太野坡',
    '越智越人',
    '河合曾良',
    '広瀬惟然',
    '服部土芳',
]

for index, haijin in enumerate(data.columns):
    wc.generate(corpus.Haiku[haijin])
    
    plt.subplot(5, 3, index+1)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.title(full_names[index])

plt.show()
```

```python
# 一人にしか使われていない単語のみを残す
data_unique = data.transpose()
for col in data_unique.columns:
    if data_unique[col].values.sum() > 1:
        data_unique = data_unique.drop(col, axis=1)

unique_dict = {}
for col in data_unique.columns:
    haijin = data_unique[data_unique[col] == 1].index[0]
    unique_dict.setdefault(haijin, 0)
    unique_dict[haijin] += 1

unique_rate_dict = {}
for haijin, u_num in unique_dict.items():
    all_word_num = len(corpus.Haiku[haijin].split(' '))
    unique_rate  = u_num / all_word_num
    unique_rate_dict[haijin] = round(unique_rate, 3)

data_unique_rate = pd.DataFrame.from_dict(unique_rate_dict, orient='index')
data_unique_rate.columns = ['unique rate']
data_unique_rate.head()
```

```python
# unique wordsの割合をbar plotで表現する

unique_rate_list = []
for haijin in data.columns:
    unique_num   = data[haijin].to_numpy().nonzero()[0].size
    all_word_num = len(corpus.Haiku[haijin].split(' '))
    unique_rate  = unique_num / all_word_num
    unique_rate_list.append(unique_rate)

data_unique = pd.DataFrame(list(zip(full_names, unique_rate_list)), columns=['haijin', 'unique rate'])
data_unique = data_unique.sort_values(by='unique rate')
data_unique
```
